## Define function to inspect every tables basic info

```python
def inspect_table(df, name):
    print("=" * 50)
    print(f"Table: {name}")
    print("=" * 50)

    print(f"Shape: {df.shape}")
    print("\nColumns:")
    print(df.columns.tolist())

    print("\nInfo:")
    df.info()

    print("\nSample:")
    display(df.head())
```

## Products with enough interaction history

```python
eligible_products = (
    product_sales[product_sales >= 5]
)

eligible_products.shape
```

```python
eligible_products.describe()
```

```python
model_df = interaction_df[
    [
        "customer_unique_id",
        "product_id",
        "product_category_name",
        "price",
        "order_purchase_timestamp"
    ]
].copy()
```

```python
model_df = model_df[
    model_df["product_id"].isin(
        eligible_products.index
    )
]
```

```python
print(model_df.shape)

print(model_df["customer_unique_id"].nunique())

print(model_df["product_id"].nunique())
```

## Finding most popular products by purchase count

```python
popular_products = (
    model_df
    .groupby("product_id")
    .size()
    .reset_index(name="purchase_count")
    .sort_values(
        "purchase_count",
        ascending=False
    )
)

popular_products.head(20)
```

```python
product_metadata = (
    interaction_df[
        [
            "product_id",
            "product_category_name"
        ]
    ]
    .drop_duplicates()
)

popular_products = popular_products.merge(
    product_metadata,
    on="product_id",
    how="left"
)
```

```python
TOP_K = 10

recommendations = (
    popular_products
    .head(TOP_K)
)

recommendations
```

## Storing in parquet format

```python
recommendations.to_parquet(
    "../artifacts/popularity_recommendations.parquet",
    engine="pyarrow",
    index=False
)
```

```python
popular_products = (
    model_df
    .groupby("product_id")
    .size()
    .reset_index(name="purchase_count")
    .sort_values(
        "purchase_count",
        ascending=False
    )
)

popular_products.head(20)
```

```python
popular_products = (
    popular_products
    .merge(
        product_df[
            [
                "product_id",
                "product_category_name"
            ]
        ],
        on="product_id",
        how="left"
    )
)
```

```python
popular_products.head(20)
```

```python
popular_products["popularity_score"] = (
    popular_products["purchase_count"]
    /
    popular_products["purchase_count"].max()
)
```

```python
popular_products.head(20)
```

```python
popular_products.to_parquet(
    "../artifacts/popularity_products.parquet",
    index=False
)
```
